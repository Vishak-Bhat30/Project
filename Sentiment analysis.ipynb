```python
# importing the libraries
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import torch.nn as nn 
```

```python
device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# importing the dataset and then extracting the data and the target
data=pd.read_csv('/kaggle/input/prometeo23-pre-event-ml/train.csv')
Y=data['label']
X=data.drop(['label'],axis=1)
```

```python
# converting to torch
X=np.array(X)
Y=np.array(Y)
X= torch.Tensor(X)
```

```python
# extract the dimensions
m,n =X.shape
```

```python
sc=StandardScaler()
X_train, X_test, y_train ,y_test= train_test_split(X,Y,test_size=0.2,shuffle=True)
```

```python
# scaling the data
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
# lets set the hyper parameters
num_epochs=300
lr=0.007
n_hidden_layers=300
n_output_layers=5
```

```python
# architecture
class NeuralNetwork(nn.Module):
    def __init__(self,n,hidden,output):
        super(NeuralNetwork,self).__init__()
        self.l1=nn.Linear(n,hidden)
        self.l2=nn.Linear(hidden,output)
        self.relu=nn.ReLU()
        
    def forward(self,X):
        out=self.l1(X)
        out=self.relu(out)
        out=self.l2(out)
        return out
        
```

```python
model=NeuralNetwork(n,n_hidden_layers,n_output_layers)
```

```python
# loss and the optimizer
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(model.parameters(),lr=lr)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size =2)
```

```python
X_train=torch.tensor(X_train,dtype=torch.float32)
X_test=torch.tensor(X_test,dtype=torch.float32)
y_train=torch.tensor(y_train,dtype=torch.float32)
y_test=torch.tensor(y_test,dtype=torch.float32)
```

```python
y_train=y_train.view(X_train.shape[0])
y_test=y_test.view(X_test.shape[0])
```

```python
for epoch in range(num_epochs):
    
    y_pred=model(X_train)
    
    loss=criterion(y_pred,y_train.type(torch.long))
    
    
    optimizer.zero_grad()
    loss.backward()
    
    optimizer.step()
    scheduler.step()
    
    if epoch%10==0:
        print(loss)
    
```
